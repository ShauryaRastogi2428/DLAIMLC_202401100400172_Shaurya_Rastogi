```python
# Step 1: Given Data

# Inputs
x1 = 75
x2 = 65
x3 = 80

# Initial weights
w1 = 4
w2 = 3
w3 = 5

# Bias
b = -1000

# Learning rate
lr = 0.01

# Target output
target = 1   # Pass = 1


# Step 2: Compute Net Input (z)
z = w1*x1 + w2*x2 + w3*x3 + b
print("Initial z =", z)

# Step 3: Activation function
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Prediction before learning =", y_pred)


# Step 4: Calculate Error
error = target - y_pred
print("Error =", error)


# Step 5: Update weights (Perceptron learning rule)
w1 = w1 + lr * error * x1
w2 = w2 + lr * error * x2
w3 = w3 + lr * error * x3

print("Updated weights:", w1, w2, w3)


# Step 6: Update bias
b = b + lr * error
print("Updated bias:", b)


# Step 7: Recalculate z
z = w1*x1 + w2*x2 + w3*x3 + b
print("New z =", z)

if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Prediction after learning =", y_pred)
```
